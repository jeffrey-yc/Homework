```python
from bs4 import BeautifulSoup
import requests
```

```python
source = requests.get('https://movies.yahoo.com.tw/movieinfo_main.html/id=5644').text
soup = BeautifulSoup(source, 'lxml')
print(soup.prettify())
```

```python
title = soup.find('title').text.replace('- Yahoo奇摩電影',' ')
print(title)
```

```python
title_cn = soup.find('div', class_='movie_intro_info_r').h1.text
print(title_cn)
```

```python
title_en = soup.find('div', class_='movie_intro_info_r').h3.text
print(title_en)
```

```python
date = soup.find('div', class_='movie_intro_info_r').span.text
print(date)
```

```python
genre= []
for movie_genre in soup.find_all('div', class_='level_name'):
    for t in movie_genre.find_all('a', class_='gabtn'):
        genre.append(t.text.replace('\n', " ").strip())
genre
```

```python
movie_info = []
for span in soup.find_all('div', class_='movie_intro_info_r'):
    for text in span.find_all('span'):
        movie_info.append(text.text)
length = movie_info [0]
date = movie_info[1]
company = movie_info[2]
print(length, date, company)
```

```python
cast_list = []
for cast in soup.find_all('div', class_='movie_intro_list'):
    if len(cast.text) == 4:
        director = cast.text
    else:
        for actor in cast.find_all('a'):
            cast_list.append(actor.text)
director = director.replace("'\n", ' ').replace("\n", " ").lstrip()
print(cast_list)
print(director)
```

```python
for web in soup.find_all('div', class_='movie_intro_info_r'):
    website = web.find('a',class_="gabtn")['href']
    print(website)
        
```

```python
summary = soup.find('div',class_='gray_infobox_inner').span.text.lstrip()
print(summary)
```

```python
import csv
csv_file = open('webscrap_yahoo.csv', 'w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Title_EN','Date','Genre','Length','Director','Cast','Production','Website','Plot']) 
csv_writer.writerow([title_cn])
csv_file.close()
```

```python
my_dict = {'Title':[title],'Date':[date],'Genre':[genre],'Length':[length], 'Company': [company], 'Director':[director], 'Cast':[cast_list],
          'Website':[website],'Summary':[summary]}
import pandas as pd
df = pd.DataFrame(my_dict)
df
```
